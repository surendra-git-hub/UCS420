```python
import numpy as np
import pandas as pd

rn = 102317040
np.random.seed(rn)

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
categories = ["Electronics", "Clothing", "Home & Kitchen", "Sports"]

sales_data = np.random.randint(1000, 5001, size=(12, 4))
df = pd.DataFrame(sales_data, columns=categories, index=months)

print("First 5 Rows of Sales Data:")
print(df.head(5))
print("\nSummary Statistics:")
print(df.describe())

totals = df.sum()
print("\nTotal Sales by Category:")
print(totals)

totalm = df.sum(axis=1)
print("\nMonthly Total Sales:")
print(totalm)

growth_rates = df.pct_change().mean(axis=0) * 100
print("\nAverage Growth Rates (%):")
print(growth_rates)

df["Total Sales"] = totalm
df["Growth Rate (%)"] = df["Total Sales"].pct_change() * 100

print("\nDataFrame with Total Sales and Growth Rate:")
print(df)

discounted_df = df.copy()
if rn % 2 == 0:
    discounted_df["Electronics"] *= 0.9
else:
    discounted_df["Clothing"] *= 0.85

print("\nDiscounted DataFrame:")
print(discounted_df)

df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
df.plot(kind='line',marker='o',title='Monthly Sales Trends')

plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend(loc='upper left')
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(data=df[categories])
plt.title("Sales Distribution by Category")
plt.show()
```
